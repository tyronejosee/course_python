## **5. Funciones integradas 煤tiles**
Python tiene varias funciones integradas que apoyan el paradigma funcional:

### **`map()`**
Aplica una funci贸n a cada elemento de un iterable.
```python
numeros = [1, 2, 3]
cuadrados = map(lambda x: x**2, numeros)
print(list(cuadrados))  # Resultado: [1, 4, 9]
```

### **`filter()`**
Filtra los elementos de un iterable seg煤n una condici贸n.
```python
numeros = [1, 2, 3, 4]
pares = filter(lambda x: x % 2 == 0, numeros)
print(list(pares))  # Resultado: [2, 4]
```

### **`reduce()`**
Acumula un resultado aplicando una funci贸n de dos argumentos. (Requiere importar `functools`).
```python
from functools import reduce

numeros = [1, 2, 3, 4]
suma_total = reduce(lambda x, y: x + y, numeros)
print(suma_total)  # Resultado: 10
```

### **`sorted()`**
Ordena elementos con un criterio definido.
```python
numeros = [3, 1, 4, 2]
ordenados = sorted(numeros, key=lambda x: x)
print(ordenados)  # Resultado: [1, 2, 3, 4]
```

## **8. Iteradores y Generadores**
Los iteradores y generadores se usan mucho en programaci贸n funcional para trabajar con secuencias de datos de manera eficiente.

### **Ejemplo de un generador:**
```python
def numeros_pares(limite):
    for i in range(limite):
        if i % 2 == 0:
            yield i

print(list(numeros_pares(10)))  # Resultado: [0, 2, 4, 6, 8]
```
