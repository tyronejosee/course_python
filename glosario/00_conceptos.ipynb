# **Conceptos**

| Concepto                       | Descripción                                                                                          |
|---------------------------------|------------------------------------------------------------------------------------------------------|
| `Algoritmo`                     | Un conjunto de instrucciones paso a paso para resolver un problema o realizar una tarea.             |
| `Array`                         | Estructura de datos que almacena elementos de un mismo tipo en una secuencia.                         |
| `Asignación`                    | El proceso de asignar un valor a una variable.                                                        |
| `Atributo`                      | Característica o propiedad de un objeto en la programación orientada a objetos.                       |
| `Bucle`                         | Estructura de control que repite un bloque de código varias veces.                                   |
| `Clase`                         | Plantilla o estructura que define propiedades y comportamientos de los objetos en programación orientada a objetos. |
| `Compilador`                    | Herramienta que traduce el código fuente de un lenguaje de programación a código de máquina.         |
| `Condicional`                   | Estructura de control que ejecuta código dependiendo de si se cumple o no una condición.              |
| `Constante`                     | Valor que no cambia durante la ejecución del programa.                                                |
| `Constructor`                   | Método especial en una clase que se ejecuta automáticamente cuando se crea un objeto de esa clase.   |
| `Depuración`                    | Proceso de encontrar y corregir errores en el código de un programa.                                 |
| `Desarrollo Ágil`               | Metodología de desarrollo de software que promueve la entrega continua de software funcional.         |
| `Encapsulamiento`               | Técnica de programación orientada a objetos que oculta los detalles internos de un objeto y expone solo lo necesario. |
| `Error`                         | Problema que ocurre en un programa que impide que funcione correctamente.                           |
| `Excepción`                     | Evento que ocurre durante la ejecución de un programa y que interrumpe el flujo normal de ejecución. |
| `Flujo de Control`              | Dirección que sigue la ejecución de un programa.                                                    |
| `Framework`                     | Conjunto de herramientas, bibliotecas y convenciones para facilitar el desarrollo de software.        |
| `Función`                       | Bloque de código que realiza una tarea específica y puede devolver un valor.                         |
| `Herencia`                      | Concepto de programación orientada a objetos donde una clase hereda características de otra clase.   |
| `IDLE`                          | Entorno de desarrollo integrado para lenguajes de programación.                                                         |
| `Implementación`                | Proceso de escribir el código para que una función o clase haga lo que se espera.                    |
| `Importar`                      | Incluir bibliotecas externas o módulos dentro de un programa para usar sus funcionalidades.          |
| `Indentación`                   | Espacios o tabulaciones que se usan para estructurar y organizar el código, especialmente en Python.  |
| `Iterador`                      | Objeto que permite recorrer una colección de datos, como una lista, de uno en uno.                   |
| `Lógica Booleana`               | Rama de la matemática y la lógica que usa valores de verdadero o falso para controlar el flujo del programa. |
| `Loop`                          | Ver `Bucle`.                                                                                         |
| `Método`                        | Función asociada a un objeto o clase en programación orientada a objetos.                             |
| `Módulo`                        | Archivo que contiene código Python, como funciones y variables, que pueden ser utilizados en otros programas. |
| `Null`                          | Valor especial que indica la ausencia de valor o que no existe ningún objeto o referencia.            |
| `Objeto`                        | Instancia de una clase en programación orientada a objetos, que contiene datos y métodos.            |
| `Operador`                      | Símbolo que realiza una operación en uno o más operandos.                                            |
| `Optimización`                  | Proceso de hacer que un programa funcione de manera más eficiente, en términos de tiempo o espacio.   |
| `Polimorfismo`                  | Concepto en programación orientada a objetos que permite que una función o método tenga diferentes comportamientos. |
| `Proyecto`                      | Conjunto de archivos, programas y configuraciones que se desarrollan para crear una aplicación.      |
| `Recursividad`                  | Técnica de programación en la que una función se llama a sí misma para resolver un problema.        |
| `Referencia`                    | El vínculo entre una variable y la ubicación de un valor en memoria.                                 |
| `Repositorio`                   | Almacenamiento de código fuente, generalmente en un sistema de control de versiones como Git.         |
| `Resolución`                    | El proceso de encontrar una solución para un problema o una tarea en programación.                   |
| `Return`                        | Palabra clave que se usa para devolver un valor desde una función.                                   |
| `Runtime`                       | El periodo en que un programa está siendo ejecutado.                                                 |
| `SQL`                           | Lenguaje de consulta estructurado utilizado para interactuar con bases de datos.                     |
| `Sintaxis`                      | Conjunto de reglas que define cómo se deben escribir las instrucciones en un lenguaje de programación. |
| `Subclase`                      | Clase que hereda de otra clase (superclase).                                                         |
| `Variable`                      | Espacio en memoria donde se almacena un valor que puede cambiar durante la ejecución del programa.   |
| `Virtualización`                | Creación de una versión virtual de algo, como servidores o sistemas operativos.                      |
| `Abstracción`                   | Técnica de ocultar los detalles complejos y mostrar solo lo necesario para interactuar con un sistema. |
| `Archivo`                       | Conjunto de datos almacenados en un dispositivo de almacenamiento que puede ser leído o modificado.    |
| `Argumento`                     | Valor que se pasa a una función o método para ser utilizado dentro de su ejecución.                   |
| `Asincronía`                    | Técnica que permite la ejecución de múltiples tareas al mismo tiempo sin bloquear el flujo del programa. |
| `Automatización`                | Uso de tecnología para realizar tareas repetitivas de forma automática.                             |
| `Big-O`                         | Notación que describe el rendimiento o la complejidad algorítmica de un programa, específicamente su tiempo de ejecución. |
| `Bug`                           | Error o defecto en el código que causa un comportamiento inesperado o no deseado en el programa.     |
| `Callback`                      | Función que se pasa como argumento a otra función y que se ejecuta después de que la función principal termine. |
| `Constantes Globales`           | Variables que se definen una vez y que pueden ser utilizadas en cualquier parte del programa.        |
| `Contenedor`                    | Objeto o estructura que puede almacenar múltiples elementos, como listas o matrices.                 |
| `Control de Versiones`          | Sistema que permite llevar un seguimiento de los cambios realizados en el código fuente a lo largo del tiempo. |
| `Corte`                         | Interrupción o detención de un proceso o programa en ejecución.                                      |
| `Cualificación Completa`        | Método de especificar la ruta completa de un objeto o clase en un entorno de programación jerárquico.  |
| `Deadlock`                       | Situación donde dos o más procesos están esperando indefinidamente por recursos que están siendo bloqueados mutuamente. |
| `Definición`                    | Proceso de establecer el valor de una variable, función u objeto en un programa.                     |
| `Encapsulación`                 | Técnica de programación que restringe el acceso directo a algunos de los componentes de un objeto.    |
| `Estructura de Datos`           | Forma de organizar y almacenar datos de manera que sean eficientes de usar y manipular.              |
| `Excepción Controlada`          | Tipo de excepción que es anticipada y gestionada explícitamente en el código.                        |
| `GIT`                           | Sistema de control de versiones distribuido que permite a los desarrolladores colaborar en el desarrollo de software. |
| `Herencia Múltiple`             | Capacidad de una clase para heredar de más de una clase base.                                        |
| `Immutabilidad`                 | Propiedad de un objeto que no puede ser modificado una vez que ha sido creado.                        |
| `Interfaz`                      | Contrato que define los métodos que una clase debe implementar, sin proporcionar su implementación.    |
| `Iterador`                      | Objeto que permite recorrer los elementos de una colección, como listas o tuplas, uno por uno.       |
| `Lenguaje de Bajo Nivel`        | Lenguaje de programación cercano al lenguaje máquina, como ensamblador.                              |
| `Lenguaje de Alto Nivel`        | Lenguaje de programación más cercano al lenguaje humano, como Python o Java.                         |
| `Manejo de Archivos`            | Técnicas y funciones para leer, escribir, y manipular archivos dentro de un programa.                 |
| `Manejo de Memoria`             | Técnicas para gestionar la memoria durante la ejecución de un programa, como la recolección de basura. |
| `Mutable`                       | Tipo de objeto cuyo valor puede ser modificado después de su creación.                                |
| `Patrón de Diseño`              | Solución general a un problema recurrente en el diseño de software.                                  |
| `Polimorfismo`                  | Propiedad de un lenguaje de programación que permite que un objeto se comporte de múltiples maneras. |
| `Puntero`                       | Variable que almacena la dirección de memoria de otra variable.                                      |
| `Rendimiento`                   | Medida de la eficiencia de un programa en cuanto a tiempo de ejecución y uso de recursos.            |
| `Repositorio de Código`         | Plataforma o sistema donde se almacena y gestiona el código fuente de un proyecto de software.       |
| `Refactorización`               | Proceso de reestructurar un código para mejorar su legibilidad o eficiencia sin cambiar su funcionalidad. |
| `Reflección`                    | Técnica que permite examinar y modificar la estructura de un programa durante su ejecución.          |
| `Renderización`                 | Proceso de convertir datos en una forma visual, como mostrar un sitio web en un navegador.           |
| `Recursión`                     | Función que se llama a sí misma para resolver un problema.                                           |
| `Segmentación`                  | División de un proceso o tarea en varias partes para su ejecución paralela o más eficiente.          |
| `Sistema Operativo`             | Software que gestiona el hardware de la computadora y proporciona servicios para programas de software. |
| `Variable Local`                | Variable que solo es accesible dentro de una función o bloque de código específico.                  |
| `Variable Global`               | Variable que puede ser utilizada y modificada en cualquier parte del programa.                      |
