## **Matplotlib**

## **Gráficos de Líneas (Line Plot)**

Se utilizan para mostrar datos continuos a lo largo de un intervalo o tiempo.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)
plt.title("Gráfico de Líneas")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

## **Gráficos de Barras (Bar Chart)**

Comparan categorías o grupos.

```python
x = ["A", "B", "C", "D"]
y = [5, 7, 3, 8]

plt.bar(x, y, color="blue")
plt.title("Gráfico de Barras")
plt.xlabel("Categorías")
plt.ylabel("Valores")
plt.show()
```

## **Gráficos de Barras Horizontales (Horizontal Bar Chart)**

Son similares a los gráficos de barras, pero las barras están dispuestas horizontalmente.

```python
plt.barh(x, y, color="red")
plt.title("Gráfico de Barras Horizontales")
plt.xlabel("Valores")
plt.ylabel("Categorías")
plt.show()
```

## **Histogramas (Histogram)**

Muestran la distribución de un conjunto de datos.

```python
data = [5, 7, 3, 8, 8, 5, 6, 8, 9, 10]

plt.hist(data, bins=5, color="purple")
plt.title("Histograma")
plt.xlabel("Intervalos")
plt.ylabel("Frecuencia")
plt.show()
```

## **Gráficos de Dispersión (Scatter Plot)**

Representan la relación entre dos variables.

```python
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

plt.scatter(x, y, color="red")
plt.title("Gráfico de Dispersión")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

## **Gráficos de Torta (Pie Chart)**

Muestran proporciones como un círculo dividido en sectores.

```python
sizes = [15, 30, 45, 10]
labels = ["A", "B", "C", "D"]
colors = ["gold", "yellowgreen", "lightcoral", "lightskyblue"]

plt.pie(sizes, labels=labels, colors=colors, autopct="%1.1f%%")
plt.title("Gráfico de Torta")
plt.axis("equal")  # Para que el círculo se vea bien proporcionado
plt.show()
```

## **Gráficos de Área (Stacked Area Plot)**

Muestran tendencias de datos acumulativos a lo largo del tiempo.

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]
y2 = [2, 3, 4, 5, 6]

plt.fill_between(x, y1, color="skyblue", alpha=0.4)
plt.fill_between(x, y2, color="orange", alpha=0.6)
plt.title("Gráfico de Área")
plt.show()
```

## **Gráficos de Caja (Box Plot)**

Muestran la distribución estadística de un conjunto de datos.

```python
data = [7, 8, 9, 5, 7, 6, 8, 10, 12, 4]

plt.boxplot(data)
plt.title("Gráfico de Caja")
plt.show()
```

## **Gráficos de Calor (Heatmap)**

Usados para visualizar matrices o relaciones de valores.

```python
import numpy as np

data = np.random.rand(6, 6)

plt.imshow(data, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.title("Gráfico de Calor")
plt.show()
```

## **Gráficos de Radar (Spider Chart)**

Comparan múltiples variables en forma de un polígono.

```python
from math import pi

categories = ['A', 'B', 'C', 'D']
values = [4, 3, 2, 5]

angles = [n / float(len(categories)) * 2 * pi for n in range(len(categories))]
angles += angles[:1]

values += values[:1]

plt.figure(figsize=(6, 6))
ax = plt.subplot(111, polar=True)
plt.xticks(angles[:-1], categories)

ax.plot(angles, values, linewidth=2, linestyle='solid')
ax.fill(angles, values, color='blue', alpha=0.4)
plt.title("Gráfico de Radar")
plt.show()
```

## **Gráficos 3D**

Muestran datos en tres dimensiones.

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]
z = [1, 3, 5, 7, 9]

ax.scatter(x, y, z)
ax.set_title("Gráfico 3D")
plt.show()
```
