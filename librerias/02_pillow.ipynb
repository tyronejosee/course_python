## **Funciones Principales**

````python
from IPython.display import display
````

- `Image.open()` Abre una imagen desde un archivo.

````python
from PIL import Image

img = Image.open("../assets/img/variable.jpg")
# img.show() # Visor predeterminado

display(img)
````

- `Image.save()` Guarda una imagen en un archivo.

````python
from PIL import Image

img = Image.open("../assets/img/variable.jpg")
img.save("../assets/img/variable2.jpg")
````

- `Image.resize()` Cambia el tamaño de una imagen.

````python
new_size = (100, 100)
resized_img = img.resize(new_size)
# resized_img.show()

display(resized_img)
````

- `Image.rotate()` Rota una imagen.

````python
rotated_img = img.rotate(45)
# rotated_img.show()

display(rotated_img)
````

- `Image.crop()` Recorta una imagen.

````python
crop_area = (50, 50, 200, 200)  # (left, upper, right, lower)
cropped_img = img.crop(crop_area)
# cropped_img.show()

display(cropped_img)
````

- `Image.convert()` Convierte la imagen a otro modo (por ejemplo, escala de grises).

````python
grayscale_img = img.convert("L")

display(grayscale_img)
````

- `Image.filter()` Aplica un filtro a una imagen.

````python
from PIL import ImageFilter

blurred_img = img.filter(ImageFilter.BLUR)

display(blurred_img)
````

- `Image.thumbnail()` Crea una miniatura de la imagen.

````python
thumbnail_size = (48, 48)
img.thumbnail(thumbnail_size)

display(img)
````

- `Image.paste()` Pega una imagen sobre otra.

````python
fondo = Image.open("../assets/img/variable.jpg")
elemento = Image.open("../assets/img/function.jpg").resize((150, 150))

fondo.paste(elemento, (50, 50))

display(fondo)
````

- `ImageDraw.Draw()` Dibuja sobre una imagen.

````python
from PIL import Image, ImageDraw

imagen2 = Image.open("../assets/img/indentation.jpg")
# img.show() # Visor predeterminado

draw = ImageDraw.Draw(imagen2)
draw.rectangle((50, 50, 200, 150), outline="red", width=6)

display(imagen2)
````

- `ImageFont.truetype()` Carga una fuente personalizada para agregar texto a una imagen.

````python
from PIL import Image, ImageFont, ImageDraw

blue_img = Image.open("../assets/img/blue.png")
draw = ImageDraw.Draw(blue_img)
fuente = ImageFont.truetype("arial.ttf", 20)
draw.text((50, 50), "Biblioteca Pillow", fill="red", font=fuente)

display(blue_img)
````

- `Image.alpha_composite()` Combina imágenes con un canal alfa.

````python
img1 = Image.open("../assets/img/blue.png").convert("RGBA")
img2 = Image.open("../assets/img/red.png").convert("RGBA")

composite = Image.alpha_composite(img1, img2)
display(composite)
````

- `Image.blend()` Combina dos imágenes con una transparencia especificada.

````python
img1 = Image.open("../assets/img/blue.png").convert("RGBA")
img2 = Image.open("../assets/img/red.png").convert("RGBA")

blended = Image.blend(img1, img2, alpha=0.5)
display(blended)
````

- `Image.transpose()` Realiza transformaciones simples, como reflejo o rotación.

````python
func_img = Image.open("../assets/img/function.jpg")

flipped_img = func_img.transpose(Image.FLIP_LEFT_RIGHT)
display(flipped_img)
````

- `Image.getdata()` Obtiene los datos de píxeles de la imagen.

````python
pixels = list(img.getdata())
print(pixels[:10])  # Imprime los primeros 10 píxeles
````

- `ImageDraw.Draw.text()` Agrega texto a una imagen.

````python
red_img = Image.open("../assets/img/red.png")
draw = ImageDraw.Draw(red_img)
draw.text((105, 128), "Text Overlay", fill="white")

display(red_img)
````

- `Image.new()` Crea una nueva imagen en blanco.

````python
blank_img = Image.new("RGB", (200, 200), color="white")
display(blank_img)
````
