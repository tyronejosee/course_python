```python
import matplotlib.pyplot as plt
import pandas as pd

# Crear un DataFrame
data = {
    'Ficha': [
        "11.493.733-9 CARLOS ALFONSO SANHUEZA PEREIRA",
        "12.920.283-1 JUAN RODRIGO FERRADA SILVA",
        "14.374.425-6 PEDRO KARIN CONTRERAS LAGOS",
        "16.152.425-5 CARLOS ALBERTO RAMIREZ ITURRA",
        "17.075.608-8 ERICK MADIEL VERGARA GALVEZ",
        "50.340.420-6 SANCHEZ Y CIA. LTDA."
    ],
    'Abonos': [297500, 71400, 44090, 168081, 193, 3344518],
    'Cargos': [0, 0, 131876, 0, 1844500, 131876]
}

df = pd.DataFrame(data)

# Gráfico de barras agrupadas
plt.figure(figsize=(10, 6))
x = range(len(df['Ficha']))
plt.bar(x, df['Abonos'], width=0.4, label='Abonos', color='skyblue', align='center')
plt.bar([i + 0.4 for i in x], df['Cargos'], width=0.4, label='Cargos', color='salmon', align='center')

# Configuración de etiquetas
plt.xlabel('Ficha')
plt.ylabel('Montos ($)')
plt.title('Comparación de Abonos y Cargos por Ficha')
plt.xticks([i + 0.2 for i in x], df['Ficha'], rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```
